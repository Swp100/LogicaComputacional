```python
# ((𝑃 → 𝑄)∧(𝑅 →𝑆)∧(¬𝑄∨¬𝑆)→(¬𝑃∨¬𝑅) 

# Cabeçalho
print(f'P | Q | R | S | 𝑃→𝑄 | 𝑅→𝑆 | (𝑃→𝑄)∧(𝑅→𝑆) | ¬𝑄∨¬𝑆 | ¬𝑃∨¬𝑅 | (¬𝑄∨¬𝑆)→(¬𝑃∨¬𝑅) | ((𝑃 → 𝑄)∧(𝑅 →𝑆)∧(¬𝑄∨¬𝑆)→(¬𝑃∨¬𝑅)')

#Testes lógicos
for a in range(0,2,1):
    for b in range(0,2,1):
        for c in range(0,2,1):
            for d in range(0,2,1):
                if a==0:
                    P="V"
                    NP="F"
                else:
                    P="F"
                    NP="V"
                if b==0:
                    Q="V"
                    NQ="F"
                else:
                    Q="F"
                    NQ="V"
                if c==0:
                    R="V"
                    NR="F"
                else:
                    R="F"
                    NR="V"
                if d==0:
                    S="V"
                    NS="F"
                else:
                    S="F"
                    NS="V"

                # Implicação 𝑃→𝑄
                if P=="V" and Q=="F":
                    T= "F"
                else:
                    T="V"

                # Implicação 𝑅→𝑆
                if R=="V" and S=="F":
                    U= "F"
                else:
                    U="V" 

                # Conjunção (𝑃→𝑄)∧(𝑅→𝑆)
                if T=="V" and U=="V":
                    PQRS="V"
                else:
                    PQRS="F"

                # Disjunção ¬𝑄∨¬𝑆
                if NQ=="V" or NS=="V":
                    NQS="V"
                else:
                    NQS="F"

                #Disjunção ¬𝑃∨¬𝑅
                if NP=="V" or NR=="V":
                    NPR="V"
                else:
                    NPR="F"

                # Implicação (¬𝑄∨¬𝑆)→(¬𝑃∨¬𝑅)
                if NQS=="V" and NPR=="F":
                    QSPR="F"
                else:
                    QSPR="V"

                if PQRS=="V" and QSPR=="V":
                    print(f'{P} | {Q} | {R} | {S} |  {T}  |  {U}  |       {PQRS}      |   {NQS}   |    {NPR}   |        {QSPR}        |                 V')
                else:
                    print(f'{P} | {Q} | {R} | {S} |  {T}  |  {U}  |       {PQRS}      |   {NQS}   |    {NPR}   |        {QSPR}        |                 F')



                    


                    
                    
```
