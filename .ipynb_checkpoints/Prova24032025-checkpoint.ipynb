```python
# ((ğ‘ƒ â†’ ğ‘„)âˆ§(ğ‘… â†’ğ‘†)âˆ§(Â¬ğ‘„âˆ¨Â¬ğ‘†)â†’(Â¬ğ‘ƒâˆ¨Â¬ğ‘…) 

# CabeÃ§alho
print(f'P | Q | R | S | ğ‘ƒâ†’ğ‘„ | ğ‘…â†’ğ‘† | (ğ‘ƒâ†’ğ‘„)âˆ§(ğ‘…â†’ğ‘†) | Â¬ğ‘„âˆ¨Â¬ğ‘† | Â¬ğ‘ƒâˆ¨Â¬ğ‘… | (Â¬ğ‘„âˆ¨Â¬ğ‘†)â†’(Â¬ğ‘ƒâˆ¨Â¬ğ‘…) | ((ğ‘ƒ â†’ ğ‘„)âˆ§(ğ‘… â†’ğ‘†)âˆ§(Â¬ğ‘„âˆ¨Â¬ğ‘†)â†’(Â¬ğ‘ƒâˆ¨Â¬ğ‘…)')

#Testes lÃ³gicos
for a in range(0,2,1):
    for b in range(0,2,1):
        for c in range(0,2,1):
            for d in range(0,2,1):
                if a==0:
                    P="V"
                    NP="F"
                else:
                    P="F"
                    NP="V"
                if b==0:
                    Q="V"
                    NQ="F"
                else:
                    Q="F"
                    NQ="V"
                if c==0:
                    R="V"
                    NR="F"
                else:
                    R="F"
                    NR="V"
                if d==0:
                    S="V"
                    NS="F"
                else:
                    S="F"
                    NS="V"

                # ImplicaÃ§Ã£o ğ‘ƒâ†’ğ‘„
                if P=="V" and Q=="F":
                    T= "F"
                else:
                    T="V"

                # ImplicaÃ§Ã£o ğ‘…â†’ğ‘†
                if R=="V" and S=="F":
                    U= "F"
                else:
                    U="V" 

                # ConjunÃ§Ã£o (ğ‘ƒâ†’ğ‘„)âˆ§(ğ‘…â†’ğ‘†)
                if T=="V" and U=="V":
                    PQRS="V"
                else:
                    PQRS="F"

                # DisjunÃ§Ã£o Â¬ğ‘„âˆ¨Â¬ğ‘†
                if NQ=="V" or NS=="V":
                    NQS="V"
                else:
                    NQS="F"

                #DisjunÃ§Ã£o Â¬ğ‘ƒâˆ¨Â¬ğ‘…
                if NP=="V" or NR=="V":
                    NPR="V"
                else:
                    NPR="F"

                # ImplicaÃ§Ã£o (Â¬ğ‘„âˆ¨Â¬ğ‘†)â†’(Â¬ğ‘ƒâˆ¨Â¬ğ‘…)
                if NQS=="V" and NPR=="F":
                    QSPR="F"
                else:
                    QSPR="V"

                if PQRS=="V" and QSPR=="V":
                    print(f'{P} | {Q} | {R} | {S} |  {T}  |  {U}  |       {PQRS}      |   {NQS}   |    {NPR}   |        {QSPR}        |                 V')
                else:
                    print(f'{P} | {Q} | {R} | {S} |  {T}  |  {U}  |       {PQRS}      |   {NQS}   |    {NPR}   |        {QSPR}        |                 F')



                    


                    
                    
```
